```python
import pandas as pd
import matplotlib.pyplot as plt

user_stat=pd.read_csv('../new_dataset/user_stat_ratings.csv')
movie_stat=pd.read_csv('../new_dataset/movie_stat_ratings.csv')
```

```python
#Etude des votes par utilisateur

user_stat.head()
user_stat.describe()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(231)
plt.hist(user_stat.nb_votes, bins=100, range=[0,1000])
plt.title('répartition du nombre de votes par utilisateur')
plt.subplot(232)
plt.hist(user_stat.moy_votes, bins=100)
plt.title('répartition de la moy des votes par utilisateur')
plt.subplot(233)
plt.hist(user_stat.std_votes, bins=100)
plt.title('répartition de l ecart-type des votes par utilisateur')
plt.subplot(234)
plt.boxplot(user_stat.nb_votes)
plt.title('nombre de votes par utilisateur')
plt.subplot(235)
plt.boxplot(user_stat.moy_votes)
plt.title('moy des votes par utilisateur');
plt.subplot(236)
plt.boxplot(user_stat.std_votes)
plt.title('ecart-type des votes par utilisateur');
```

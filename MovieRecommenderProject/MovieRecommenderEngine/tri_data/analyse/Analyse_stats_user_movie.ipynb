```python
import pandas as pd
import matplotlib.pyplot as plt

user_stat=pd.read_csv('../../new_dataset/user_stat_ratings.csv')
movie_stat=pd.read_csv('../../new_dataset/movie_stat_ratings.csv')
movie_stat.dropna(inplace=True)
```

IMBD data

```python
#Overview of votes per user

user_stat.head()
user_stat.describe()
```

```python
ax=plt.figure(figsize=(15,10))
plt.subplot(231)
plt.hist(user_stat.nb_votes, bins=100, range=[0,1000])

plt.subplot(232)
plt.hist(user_stat.moy_votes, bins=100)

plt.subplot(233)
plt.hist(user_stat.std_votes, bins=100)

plt.subplot(234)
plt.boxplot(user_stat.nb_votes)
plt.title('number of votes per user')
plt.xticks([])
plt.subplot(235)
plt.boxplot(user_stat.moy_votes)
plt.title('average of votes per user');
plt.xticks([])
plt.subplot(236)
plt.boxplot(user_stat.std_votes)
plt.title('standart error of votes per user');
plt.xticks([]);
```

```python
import pandas as pd

df_conv=pd.read_csv('../../dataset/ml-20m/links.csv', index_col=0)
df_imdb=pd.read_csv('../new_dataset/imbd_df.csv', sep=',', index_col=0)
df_ref=pd.read_csv('../new_dataset/correspondances_transitoire_Id_movie', sep=',', index_col=0)
```

```python
df_conv.head()
df_ref.head()
```

```python
df_ref.head()
df_ref=df_ref.join(df_conv.imdbId)
df_ref["movieId"]=df_ref.index
df_ref.index=df_ref.imdbId
```

```python
df_ref.info()
```

```python
df=df_imdb.join(df_ref.movieId_2, on=df_imdb.index)
```

```python
df=df.dropna(subset=["movieId_2"])
df=df.sort_values(by="movieId_2")
df['movieId_ref']=range(1,23609)
df_ref=df_ref.join(df.movieId_ref)
```

```python
df_ref=df_ref.dropna(subset=["movieId_ref"])
df_ref.head()
```

```python
df.index=df.movieId_ref
df=df.drop(["movieId_2", "movieId_ref"], axis=1)
df.head()
```

```python
#Enregistrement du nouveau df avec nouveaux index comparables à dataset MovieLens
df.to_csv('../new_dataset/movie_df.csv', sep=',', columns=df.columns)
df_ref.to_csv('../new_dataset/correspondances_Id_movie', sep=',', columns=df_ref.columns)
```

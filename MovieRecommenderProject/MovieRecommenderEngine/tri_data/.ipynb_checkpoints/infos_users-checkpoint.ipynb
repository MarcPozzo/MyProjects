```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
ratings=pd.read_csv('../../dataset/ml-20m/ratings.csv')
```

```python
#Création d'une colonne avec temps entre 2 votes
ratings=ratings.sort_values(by=["userId","timestamp"])
ratings = ratings.reset_index(drop=True)
time=list(ratings.timestamp)
diff_time=[0]
for i in range(1,len(time)):
    diff_time.append((time[i]-time[i-1])/360)

ratings["time_btw_vote"]=diff_time

#liste des indices ou changement de user Id:
liste_indice=[]
temp=ratings
n_user=ratings.userId.unique().shape[0]
kmin=0
```

```python
index_users = ratings['userId'].value_counts().sort_index()
index_users = np.cumsum(index_users)
```

```python
ratings['time_btw_vote'] = ratings['timestamp'].diff()

ratings['time_btw_vote'][index_users[:-1]] = 0
ratings['time_btw_vote'][0] = 0
```

```python
#Création un df avec toutes les infos users

#Chargement de movie_df
#Suppression des colonnes inutiles, ou string
df_movie=pd.read_csv('../new_dataset/movie_df.csv', index_col=0, na_values=['', '\\N'])
to_drop=["tconst","originalTitle","primaryTitle","adult","directors","writers","actors"]
df_movie=df_movie.drop(to_drop,axis=1)
df_movie.head()
```

```python
df_movie = df_movie.fillna(df_movie.mean())
```

```python
df_movie.info()
```

```python
df_ref=pd.read_csv('../new_dataset/correspondances_Id_movie', index_col=3)
df_ref.head()
```

```python
ratings.head()
```

```python
ratings=ratings.join(df_ref.movieId_ref)
```

```python
ratings=ratings.drop(["timestamp", 'movieId'], axis=1)
ratings=ratings.dropna(subset=["movieId_ref"])
```

```python
ratings.head()
```

```python
print(len(ratings.movieId_ref.unique()))
ratings.movieId_ref=ratings.movieId_ref.astype(int)
print(ratings.shape)
ratings.head()

df=ratings.merge(right=df_movie, how='outer',left_on=['movieId_ref'], right_index=True)

df=df.dropna()
df.head(10)
```

```python
del ratings
del df_ref
del df_movie
del time
del diff_time
del liste_indice
del temp
del n_user
```

```python
#Création de nouvelles colonnes avec moy/nb de films par user

user_avg_note=df[["userId","rating"]].groupby("userId").mean()
df=df.merge(right=user_avg_note, how='outer', left_on=['userId'], right_index=True, suffixes=('','_user_avg'))

user_nb_note=df[["userId","rating"]].groupby("userId").count()
df=df.merge(right=user_nb_note, how='outer', left_on=['userId'], right_index=True, suffixes=('','_user_nb'))


df.head()
```

```python
#Création de nouvelles colonnes avec moy/nb de films par user par genre

genres_films=['action', 'adventure', 'animation', 'biography','comedy','crime', 'documentary', 
               'drama', 'family','fantasy', 'history', 'horror', 'romance','sci-fi', 'thriller','war', 'western']

df_element=df.drop(["movieId_ref","isAdult"], axis=1)

df_group_tot=df_element[["userId","rating"]].groupby("userId").count()

for element in tqdm(genres_films):
    df_group=df_element[["userId",element,"rating","startYear"]][df_element[element]==1]
    df_group_count=df_group[["userId",element,"rating"]].groupby(['userId']).count()
    df_group_avg=df_group[["userId",element,"rating","startYear"]].groupby(['userId']).mean()
    df_group_tot[str(element+"_nb_note_user")]=df_group_count.rating
    df_group_tot[str(element+"_avg_note_user")]=df_group_avg.rating
    df_group_tot[str(element+"_avg_year_user")]=df_group_avg.startYear

del df_group_avg
del df_group_count
del df_element
df=df.merge(right=df_group_tot, how="outer", left_on=['userId'], right_index=True, suffixes=('','2')).drop(["rating2"], axis=1)
df.head()
```

```python
df.info()
```

```python
df_user=df.groupby("userId").mean()
to_drop=["rating","time_btw_vote", "movieId_ref", "isAdult", "startYear", \
         "runtimeMinutes", "action", "adventure", "animation", "biography", \
         "comedy", "crime", "documentary", "drama", "family", "fantasy", \
         "history", "horror", "romance", "sci-fi", "thriller", "war", \
         "western", "averageRating", "numVotes"]
df_user=df_user.drop(to_drop, axis=1)
```

```python
df_note=df[["userId","rating","time_btw_vote", "movieId_ref"]]
```

```python
df_movie=df.groupby("movieId_ref").mean()
to_keep=["isAdult", "startYear", \
         "runtimeMinutes", "action", "adventure", "animation", "biography", \
         "comedy", "crime", "documentary", "drama", "family", "fantasy", \
         "history", "horror", "romance", "sci-fi", "thriller", "war", \
         "western", "averageRating", "numVotes"]
df_movie=df_movie[to_keep]
```

```python
df_user.to_csv('../../dataset/data_regression/df_user.csv', sep=',', columns=df_user.columns)
df_note.to_csv('../../dataset/data_regression/df_note.csv', sep=',', columns=df_note.columns)
df_movie.to_csv('../../dataset/data_regression/df_movie.csv', sep=',',columns=df_movie.columns)
```

```python
ratings=pd.read_csv('../../../dataset/ml-20m/ratings.csv')
df_ref=pd.read_csv('../../new_dataset/correspondances_Id_movie')
```

```python
# Ajout de movieId_ref dans fichier ratings, et suppression infos inutiles

ratings = ratings.set_index(['movieId'], drop=True)
df_ref = df_ref.set_index(['movieId'], drop=True)
ratings=ratings.join(df_ref.movieId_ref)

ratings=ratings.drop(["timestamp"], axis=1)
ratings=ratings.dropna(subset=["movieId_ref"])

print(len(ratings.movieId_ref.unique()))
ratings.movieId_ref=ratings.movieId_ref.astype(int)
ratings.head()

del df_ref
```

```python
ratings.head()
```

```python
#séparer jeu de test jeu d entrainement

train_data, test_data = train_test_split(ratings, test_size=0.2, train_size=.4, random_state=222)
```

```python
train_data.head()
```

```python
#Définition du nombre de users et du nombre de films

n_users = ratings.userId.unique().shape[0]
n_movies = ratings.movieId_ref.unique().shape[0]
print ("Nombre d'utilisateurs : " + str(n_users) + " | Nombre de films : " + str(n_movies))

del ratings
```

```python
train_data = train_data.groupby(['userId']).agg({'rating':list, 'movieId_ref':list})
```

```python
train_data.head()
```

```python
dico_train = {}
for line in tqdm(train_data.itertuples()):
    dico_train[line[0]] = dict(zip(line[2], line[1]))
train_data = dico_train 
del dico_train
```

```python
test_data = test_data.groupby(['userId']).agg({'rating':list, 'movieId_ref':list})


dico_test = {}
for line in tqdm(test_data.itertuples()):
    dico_test[line[0]] = dict(zip(line[2], line[1]))
test_data = dico_test 
del dico_test
```

```python

# 
k = 10

# liste contenant les ids des films ordonnés
movie_ranks = []
# liste contenant les similarités des films ordonnés
movie_sorted_similarities = []
# liste contenant la longueur des vecteurs de similarité (simplement pour faire des checks)
lengths = []

# Chargement matrice similarité
with open('../../../dataset/movie_similarity.csv', 'r') as file:
    
    for line in tqdm(file):
        # on parse les lignes du fichier
        similarities = np.fromstring(line, sep=' ')
        
        # on garde la longueur de la liste pour vérification
        lengths.append(len(similarities))
        
        # on ne garde que les k plus importantes 
        ranking = np.argsort(similarities)[-k:]

        movie_ranks.append(ranking)
        movie_sorted_similarities.append(similarities[ranking])
        
print(np.unique(lengths))
print(len(movie_sorted_similarities))
print(len(movie_ranks))
del lengths
```

```python
rmse = 0
n_movies = 0

# on parcourt les utilisateurs du test set
for user in tqdm(test_data.keys()):
    
    # on garde de côté la liste des films que l'utilisateur a vus et qui sont dans le train set
    seen_films = train_data.get(user, [])
    
    # on parcourt les films qu'il a vu mais qui font partie du jeu de test 
    for film_test in test_data[user].keys():
        # on incrémente le nombre de film
        n_movies += 1
        
        # on garde de côté la note qu'il a attribuée à ce film
        note_test = test_data[user][film_test]
        
        # on initialise la note du film d'après notre modèle
        note_train = 0
        
        # on initialise le terme de normalisation
        normalisation = 0.0000001
        
        # on garde la liste des films les plus similaires 
        most_similar_films = movie_ranks[film_test]
        
        # on garde la liste des similarités
        most_similar_films_similarities = movie_sorted_similarities[film_test]
        
        # on parcourt les indices de 0 à k-1
        for index_film_similar in range(k):
            
            # on regarde le kieme film le plus similaire
            similar_film = most_similar_films[index_film_similar]
            
            # on vérifie que le film est bien dans les films que l'utilisateur a déjà notés
            if similar_film in seen_films.keys():
                
                # si oui on peut ajouter a la note la note du film du train pondérée par la similarité entre les deux films
                similarity = most_similar_films_similarities[index_film_similar]
                note_train += similarity * seen_films[similar_film]
                normalisation += similarity
        note_train = note_train / normalisation
        test_data[user][film_test] = [note_test, note_train]
        rmse += (note_train-note_test)**2
        
```

# Filtrage collaboratif :
Ici movie-based filtering

```python
# Définitions des fonctions pour calcul des prédictions, 
# et calcul des RMSE

def pred_movie(matrice_modele, movie_similarity, k):    
    pred = np.zeros(matrice_modele.shape)
    for i in tqdm(range(matrice_modele.shape[1])):     
        top_k_movies = np.argsort(movie_similarity[:,i])[-1:-k-1:-1]
        for u in range(matrice_modele.shape[0]):
            mov=movie_similarity[i,top_k_movies]
            mat=matrice_modele[u,top_k_movies].toarray()
            pred[u,i]=mov.dot(mat.T)/(np.sum(np.abs(mov))+0.000001)   
            del mat
            del mov
    return pred 

def get_rmse(pred, actual):
    pred = pred[actual.nonzero()].flatten() # Ignore nonzero terms.
    actual = actual[actual.nonzero()].flatten()
    return np.sqrt(mean_squared_error(pred, actual))
```

# Filtrage collaboratif :
Ici user-based filtering

```python
#Chargement matrice similarité sur les users
user_similarity=np.loadtxt('../../../dataset/user_similarity.csv')
```

```python
#Calcul des prédictions
def pred_user(matrice_modele, user_similarity, k):    
    pred = np.zeros(matrice_modele.shape)
    print(pred.shape)
    for u in tqdm(range(matrice_modele.shape[0])):     
        top_k_users = np.argsort(user_similarity[:,u])[-1:-k-1:-1]  
        for i in tqdm(range(matrice_modele.shape[1])):
            pred[u,i]=user_similarity[u,:][top_k_users].dot(matrice_modele[:,i][top_k_users])
            pred[u,i]/=np.sum(np.abs(user_similarity[u,:][top_k_users]))+0.000001     
    return pred 


def get_rmse(pred, actual):
    pred = pred[actual.nonzero()].flatten() # Ignore nonzero terms.
    actual = actual[actual.nonzero()].flatten()
    return np.sqrt(mean_squared_error(pred, actual))
```

```python
# Script pour faire tourner algo filtrage collaboratif, sur users
#avec k-array pour trouver meilleur nombre de films à comparer
#affichage courbe RMSE train et test en fonction de k


k_array = [5, 15, 30, 50]
user_test_rmse = []
user_train_rmse = []
for k in k_array:
    user_pred = pred_user(matrice_modele, user_similarity, k=k)      
    user_train_rmse += [get_rmse(user_pred, matrice_modele) ]       
    user_test_rmse +=  [get_rmse(user_pred, matrice_test) ]


sns.set()
pal = sns.color_palette("Set1", 4)
plt.figure(figsize=(8, 8))
plt.plot(k_array, user_train_rmse, c=pal[2], label='User-based train', alpha=0.5, linewidth=2)
plt.plot(k_array, user_test_rmse, c=pal[3], label='User-based test', linewidth=2)
plt.legend(loc='best', fontsize=20)
plt.xlabel('k', fontsize=30);
plt.ylabel('RMSE', fontsize=30);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.8))
```

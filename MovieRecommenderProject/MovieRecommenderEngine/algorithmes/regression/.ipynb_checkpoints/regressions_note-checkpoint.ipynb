```python
#régression sur prédictions des notes ratings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn import model_selection, preprocessing
from sklearn.model_selection import cross_val_predict, cross_val_score, cross_validate, train_test_split
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error
from tqdm import tqdm
```

```python
#On charge dataframe movie_df_avec_users et on gère valeurs nulles
df=pd.read_csv('../../new_dataset/movie_df_avec_users.csv', sep=',')
df.head()
```

```python
to_drop=["userId", "movieId_ref"]
df=df.drop(to_drop, axis=1)
df=df.fillna(0)

```

```python
#df.startYear=preprocessing.scale(df.startYear)
#df.runtimeMinutes=preprocessing.scale(df.runtimeMinutes)
#df.numVotes=preprocessing.scale(df.numVotes)
#df.userId=preprocessing.scale(df.userId)
#df.rating=preprocessing.scale(df.rating)
#df.movieId_ref=preprocessing.scale(df.movieId_ref)
#df.averageRating=preprocessing.scale(df.averageRating)
#df.head()
```

```python
liste=[]
for column in df.columns:
    liste.append(df[column])
plt.figure(figsize=(20,20))
plt.boxplot(liste)
plt.xticks(range(1,27),df.columns, rotation=90);
```

```python
data=df.drop(["rating", "movieId_ref"], axis=1)
target=df.rating

X_train,X_test,y_train,y_test=train_test_split(data, target, test_size=0.2, random_state=222)
```

```python
#régression linéaire multiple

lr=LinearRegression()
lr.fit(X_train,y_train)
print("r2 train :\t",lr.score(X_train,y_train))
print("r2 test :\t",lr.score(X_test,y_test))
pred_train=lr.predict(X_train)
pred_test=lr.predict(X_test)
print("RMSE train :\t",np.sqrt(mean_squared_error(y_train,pred_train)))
print("RMSE test :\t",np.sqrt(mean_squared_error(y_test,pred_test)))

#RMSE 0.9, sur note [-1,1]. mais pas d'overfitting!
```

```python
# régression Ridge

alphas=[1,10,50,100,200,500,800]
lr_ridge=RidgeCV(alphas=alphas, fit_intercept=False)
lr_ridge.fit(X_train,y_train)
print("best alpha :\t",lr_ridge.alpha_)
print("r2 train :\t",lr_ridge.score(X_train,y_train))
print("r2 test :\t",lr_ridge.score(X_test,y_test))
pred_train=lr_ridge.predict(X_train)
pred_test=lr_ridge.predict(X_test)
print("RMSE train :\t",np.sqrt(mean_squared_error(y_train,pred_train)))
print("RMSE test :\t",np.sqrt(mean_squared_error(y_test,pred_test)))

#RMSE 0.90, pas d'overfitting.
```

```python
#régression Lasso

lr_lasso=LassoCV(cv=10)
lr_lasso.fit(X_train,y_train)
print("best alpha :\t",lr_lasso.alpha_)
print("r2 train :\t",lr_lasso.score(X_train,y_train))
print("r2 test :\t",lr_lasso.score(X_test,y_test))
pred_train=lr_lasso.predict(X_train)
pred_test=lr_lasso.predict(X_test)
print("RMSE train :\t",np.sqrt(mean_squared_error(y_train,pred_train)))
print("RMSE test :\t",np.sqrt(mean_squared_error(y_test,pred_test)))

#Afficher les coefficients des variables par ordre d'importance
coeff=lr_lasso.coef_
var_non_retenues=0
for k in range(coeff.shape[0]):
    if coeff[k]==0:
        var_non_retenues+=1
var_retenues=len(lr_lasso.coef_)-var_non_retenues
print('variables retenues : ', var_retenues)
print('variables non retenues : ', var_non_retenues)
pd.DataFrame({'coefs (valeurs absolues)':abs(np.round(lr_lasso.coef_,2))}, 
             index=data.columns).sort_values(by='coefs (valeurs absolues)', 
                                              ascending=False).head(10)
#RMSE 0.9, pas d'overfitting, 18 variables retenues
```

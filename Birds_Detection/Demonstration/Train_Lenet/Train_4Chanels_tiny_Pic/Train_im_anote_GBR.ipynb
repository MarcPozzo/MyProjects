```python
import numpy as np
import tensorflow
import matplotlib.pyplot as plt
from sklearn import metrics
import functions_4C as fn
```

```python
tiny_image_path='../../../../../4C_tiny_picture/'
path_to_save=Mat_path="../../../Materiels/Models"
```

```python
X_train=np.load(tiny_image_path+'imagettes4C_BGR_train.npy')
X_test=np.load(tiny_image_path+'imagettes4C_BGR_test.npy')

Y_train=np.load(tiny_image_path+'labels4C_BGR_train.npy')
Y_test=np.load(tiny_image_path+'labels4C_BGR_test.npy')
```

## 100 Epochs

```python
lenet=fn.nn()
tensorflow.random.set_seed(2)
history=lenet.fit(X_train,Y_train,validation_data=(X_test, Y_test),  epochs=100,verbose=1)
```

```python
plt.figure(figsize=(16,12))

plt.plot(history.history['accuracy'],label="train");
plt.plot(history.history['val_accuracy'],label="test");
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title("Accuracy en fonction du nombre d'epoch");
plt.legend();
```

```python
Y_predict=lenet.predict(X_test).argmax(axis=1)
print(metrics.classification_report(Y_predict,Y_test,digits=5))
lenet.save(path_to_save+'100ep_GBR_6CL_4CH')
```

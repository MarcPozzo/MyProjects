## Division de la taille de l'imagette gnre divise par la taille de l'imagette anote

```python
plt.figure(figsize=(12,12))
sns.set(); 

for i in range(5):
    rapport_taille_im_gen_anote = [int(b) / int(m) for b,m in zip(im_gen[i], im_anote[i])]
    plt.subplot(3,2,i+1)
    plt.title("dossier"+str(i))
    ax = sns.distplot(rapport_taille_im_gen_anote)
```

## BlocK Size 53

```python
path_save="/mnt/VegaSlowDataDisk/c3po_interface_mark/Materiels/listes/Comp_imgen_imanote/BS53/"
with open(path_save+"surface_carre_gen_tous_fichiers_BS53.txt", "rb") as fp:   # Unpickling
    im_gen = pickle.load(fp)
    
with open(path_save+"surface_anote_liste_tous_fichiers_BS53.txt", "rb") as fp:   # Unpickling
    im_anote = pickle.load(fp)    
    
flat_im_gen = [item for sublist in im_gen for item in sublist]
flat_im_anote = [item for sublist in im_anote for item in sublist]

plt.figure(figsize=(12,12))
sns.set(); 

for i in range(5):
    rapport_taille_im_gen_anote_ce_dossier = [int(b) / int(m) for b,m in zip(im_gen[i], im_anote[i])]
    plt.subplot(3,2,i+1)
    plt.title("dossier"+str(i))
    ax = sns.distplot(rapport_taille_im_gen_anote_ce_dossier)
    
rapport_taille_im_gen_anote_tous_dossiers = [int(b) / int(m) for b,m in zip(flat_im_gen, flat_im_anote)]

plt.subplot(3,2,6)
plt.title("tous les dossiers confondus")
ax = sns.distplot(rapport_taille_im_gen_anote_tous_dossiers)

print(len(flat_im_gen))
print("la médiane est",np.median(rapport_taille_im_gen_anote_tous_dossiers))
print("la médiane est",np.mean(rapport_taille_im_gen_anote_tous_dossiers))
```

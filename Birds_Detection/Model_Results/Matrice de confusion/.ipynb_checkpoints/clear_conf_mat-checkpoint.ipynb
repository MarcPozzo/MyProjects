```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

def get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes):
    fn=nb_oiseaux-tp
    tn=nb_imagettes-fp-fn-tp
    confusion_matrix=np.array([[tp,fp],[fn,tn]])
    return confusion_matrix

def clear_map_1(mat_conf):
    #mat_conf=get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes)

    #fig, ax = plt.subplots(figsize=(10,6))
    group_names = ["True Pos","False Pos","False Neg","True Neg"]

    group_counts = ["{0:0.0f}".format(value) for value in
                    mat_conf.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in
                         mat_conf.flatten()/np.sum(mat_conf)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]

    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]

    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(mat_conf, annot=labels, fmt="", cmap='Blues', cbar=False).set_title('matrice de confusion dossier 1');
    
    
def clear_map_2(mat_conf):
    #mat_conf=get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes)

    #fig, ax = plt.subplots(figsize=(10,6))
    group_names = ["True Pos","False Pos","False Neg","True Neg"]

    group_counts = ["{0:0.0f}".format(value) for value in
                    mat_conf.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in
                         mat_conf.flatten()/np.sum(mat_conf)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]

    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]

    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(mat_conf, annot=labels, fmt="", cmap='Blues', cbar=False).set_title('matrice de confusion dossier 2');
    
    
def clear_map_3(mat_conf):
    #mat_conf=get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes)

    #fig, ax = plt.subplots(figsize=(10,6))
    group_names = ["True Pos","False Pos","False Neg","True Neg"]

    group_counts = ["{0:0.0f}".format(value) for value in
                    mat_conf.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in
                         mat_conf.flatten()/np.sum(mat_conf)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]

    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]

    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(mat_conf, annot=labels, fmt="", cmap='Blues', cbar=False).set_title('matrice de confusion dossier 3');
    
    
def clear_map_4(mat_conf):
    #mat_conf=get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes)

    #fig, ax = plt.subplots(figsize=(10,6))
    group_names = ["True Pos","False Pos","False Neg","True Neg"]

    group_counts = ["{0:0.0f}".format(value) for value in
                    mat_conf.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in
                         mat_conf.flatten()/np.sum(mat_conf)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]

    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]

    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(mat_conf, annot=labels, fmt="", cmap='Blues', cbar=False).set_title('matrice de confusion dossier 4');    
    

def clear_map_5(mat_conf):
    #mat_conf=get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes)

    #fig, ax = plt.subplots(figsize=(10,6))
    group_names = ["True Pos","False Pos","False Neg","True Neg"]

    group_counts = ["{0:0.0f}".format(value) for value in
                    mat_conf.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in
                         mat_conf.flatten()/np.sum(mat_conf)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]

    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]

    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(mat_conf, annot=labels, fmt="", cmap='Blues', cbar=False).set_title('matrice de confusion dossier 5');    
    
    
```

```python
NB_total=[274, 930, 70, 492, 350]
TP=[59, 270, 18, 113, 75]
FP=[13, 211, 2, 29, 55]
NB_imagettes_folder=[13961, 70420, 115279, 238759, 283899]



mat_liste=[]
i=1
for nb_oiseaux,tp,fp,nb_imagettes in zip(NB_total,TP,FP,NB_imagettes_folder):
    #print("Matrice de confusion dossier", i)
    #print(get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes))
    mat_liste.append(get_confusion_matrix(nb_oiseaux,tp,fp,nb_imagettes))
    i+=1


#clear_map_1(mat_liste[0])

plt.figure(figsize=(20,16))
plt.subplot(2,3,1)
clear_map_1(mat_liste[0])
plt.subplot(2,3,2)
clear_map_2(mat_liste[1])
plt.subplot(2,3,3)
clear_map_3(mat_liste[2])
plt.subplot(2,3,4)
clear_map_4(mat_liste[3])
plt.subplot(2,3,5)
clear_map_5(mat_liste[4])

```

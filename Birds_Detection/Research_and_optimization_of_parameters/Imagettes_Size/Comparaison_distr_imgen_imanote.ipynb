```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
```

```python
path_save="/mnt/VegaSlowDataDisk/c3po_interface_mark/Materiels/listes/Comp_imgen_imanote/"

with open(path_save+"surface_carre_gen_tous_fichiers.txt", "rb") as fp:   # Unpickling
    im_gen = pickle.load(fp)
    
with open(path_save+"surface_anote_liste_tous_fichiers.txt", "rb") as fp:   # Unpickling
    im_anote = pickle.load(fp)    
    
flat_im_gen = [item for sublist in im_gen for item in sublist]
flat_im_anote = [item for sublist in im_anote for item in sublist]
```

## Division de la taille de l'imagette générée divisée par la taille de l'imagette anotée

```python
plt.figure(figsize=(12,12))
sns.set(); 

for i in range(5):
    rapport_taille_im_gen_anote = [int(b) / int(m) for b,m in zip(im_gen[i], im_anote[i])]
    plt.subplot(3,2,i+1)
    plt.title("dossier"+str(i))
    ax = sns.distplot(rapport_taille_im_gen_anote)
```
